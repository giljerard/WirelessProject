```python
import matplotlib.pyplot as plt
```

```python
# Data
labels = ['Bluetooth', 'AirDrop', 'Future P2P WiFi']
values = [0.63, 2.17, 0.11]

# Creating the bar chart
plt.figure(figsize=(8, 5))
# plt.bar(labels, values, color=['blue', 'green'])
plt.bar(labels, values)

# Adding titles and labels
plt.title('Average Power Consumption')
plt.ylabel('Power (Watts)')
plt.ylim(0, 2.5)  # Set y-axis limit to fit the highest value


plt.savefig('pwr_ther.png')
plt.show()


```

```python
# Data
labels = ['Bluetooth', 'AirDrop', 'Future P2P WiFi']
values = [880, 293000, 14000]

# Creating the bar chart
plt.figure(figsize=(8, 5))
# plt.bar(labels, values, color=['blue', 'green'])
plt.bar(labels, values)

# Adding titles and labels
plt.title('Average Data Rate')
plt.ylabel('Data rate (Kbps)')
plt.ylim(0, 15000)  # Set y-axis limit to fit the highest value

plt.text(1, 13500, '293,000 Kbps', # x, y position
         horizontalalignment='center',
         verticalalignment='center',
         fontsize=12, color='black',
         bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=1'))

plt.savefig('rate_ther.png')
plt.show()
```
